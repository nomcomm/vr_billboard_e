```python
import os, sys, random
import pandas as pd
import numpy as np
```

```python
# 5 commercial
# 5 health_neutral
# 5 tailored_match
# 5 tailored_unmatch

commercial_list = [   'commercial_neutral_1',
                      'commercial_neutral_2',
                      'commercial_neutral_3',
                      'commercial_neutral_4',
                      'commercial_neutral_5']

health_neutral_list =['health_neutral_1',
                      'health_neutral_2',
                      'health_neutral_3',
                      'health_neutral_4',
                      'health_neutral_5']


# for each of these, we will have two separate named versions, 
# e.g., matched_unmatched_drunk_driving_A and matched_unmatched_drunk_driving_C for AA and caucasia, respectively
matched_unmatched_list = [ 'matched_unmatched_1',
                      'matched_unmatched_2',
                      'matched_unmatched_3',
                      'matched_unmatched_4',
                      'matched_unmatched_5',
                      'matched_unmatched_6',
                      'matched_unmatched_7',
                      'matched_unmatched_8',
                      'matched_unmatched_9',
                      'matched_unmatched_10']

a_c_list = ['_A', '_C'] *5
a_c_list

#a_c_list_inv = 

#n_billboards = len(health_list)
```

```python
subj_file_names_A = []
for i in list(np.arange(40)+1):
     subj_file_names_A.append('sub' + (str(i).zfill(3)) + '_A_billboards_positions.csv')

subj_file_names_A[:3]
```

```python
subj_file_names_C = []
for i in list(np.arange(40)+1):
     subj_file_names_C.append('sub' + (str(i).zfill(3)) + '_C_billboards_positions.csv')

subj_file_names_C[:3]
```

```python
for sub_count in range(40):
    
    np.random.shuffle(matched_unmatched_list) #shuffle the m_u_list so that the a/c assignments vary over people
    
    bb_list = []
    bb_list.extend(commercial_list) #always use same commercials
    bb_list.extend(health_neutral_list) #always use same health
    
    for mu_count in range(10):
        curr_m_u = matched_unmatched_list[mu_count] + a_c_list[mu_count]  #combine the shuffled m/u items with the a/c/a/c -- so that there's always 5 mathced, 5 unmatched
        bb_list.append(curr_m_u)
        #print(curr_m_u)
    
    np.random.shuffle(bb_list)  #shuffle the list so that it is all random in terms of sequence ...
    

    #print out one for A
    df = pd.DataFrame()
    df['position'] = np.arange(20)+1
    df['billboard_name'] = bb_list
    df.to_csv('../data/stimuli_subject_filesT/' + subj_file_names_A[sub_count], index=False)
    print(subj_file_names_A[sub_count])
    
    #print out one for C -- same as for A, but the matching/unmatching will be opposite.
    df = pd.DataFrame()
    df['position'] = np.arange(20)+1
    df['billboard_name'] = bb_list
    df.to_csv('../data/stimuli_subject_filesT/' + subj_file_names_C[sub_count], index=False)
    print(subj_file_names_C[sub_count])
```
