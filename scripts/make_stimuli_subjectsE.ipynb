```python
import os, sys, random
import pandas as pd
import numpy as np
```

```python
health_list = ['buckle_up',         #1
               'binge_drinking',    #2
               'hiv',               #3
               'sunscreen',         #4
               'vaccination',       #5
               'distracted_driving',#6
               'underage_drinking', #7
               'drugged_driving',   #8
               'diabetes',          #9
               'smoking']           #10

commercial_list = ['brunch',            #1
                   'burger',            #2
                   'coffee',            #3
                   'education_donation',#4
                   'furniture',         #5
                   'hotel',             #6
                   'lawyer',            #7
                   'pizza',             #8
                   'salad',             #9
                   'smartphone']       #10

emo_neu_list = ['_emo',
                '_neu'] * 5

emo_neu_list_inv = ['_neu',
                    '_emo'] * 5

prefix = 'resources/billboards/'
postfix ='.bmp'

n_billboards = len(health_list)
```

```python
# say 40 participants -- everybody gets 20 messages. ...
```

```python
subj_file_names = []
for i in list(np.arange(40)+1):
     subj_file_names.append('sub' + (str(i).zfill(3)) + '_billboards_positions.csv')

subj_file_names[:3]
```

```python
health_list
```

```python
sub_count = 0

for current_subj in range(20):
    print('-----')
    print(current_subj)
    print('-----')
    
    sub_count = sub_count +1
    print(sub_count)

    current_list  = []
    current_list_inv = []

    random.shuffle(health_list)
    random.shuffle(commercial_list)

    
    
    for counter in range(n_billboards):
        current_type          = emo_neu_list[counter]
        current_type_inv      = emo_neu_list_inv[counter]
        current_health_bb     = health_list[counter] 
        current_commercial_bb = commercial_list[counter] 
        #print(current_type)
        #print(current_health_bb)
        #print(current_commercial_bb)
    
        current_list.append(prefix + current_health_bb     + current_type + postfix)
        current_list.append(prefix + current_commercial_bb + current_type + postfix)
    
        current_list_inv.append(prefix + current_health_bb     + current_type_inv + postfix)
        current_list_inv.append(prefix + current_commercial_bb + current_type_inv + postfix)
    
    #complicated/more than it has to be function to sort both lists the same way
    sorting_array = np.arange(20)

    np.random.shuffle(sorting_array)
    print(sorting_array)
    arrayed_current_list = np.asarray(current_list)
    arrayed_current_list_sorted = arrayed_current_list[sorting_array]

    arrayed_current_list_inv = np.asarray(current_list_inv)
    arrayed_current_list_sorted_inv = arrayed_current_list_inv[sorting_array]


    
    
    df = pd.DataFrame()
    df['position'] = np.arange(20)+1
    df['billboard_name'] = arrayed_current_list_sorted
    df.to_csv('../data/stimuli_subject_files/' + subj_file_names[sub_count-1], index=False)
    print(arrayed_current_list_sorted[:1])
    print(subj_file_names[sub_count-1])
        
    print('---')
    sub_count = sub_count +1
    print(sub_count)
    df['billboard_name'] = arrayed_current_list_sorted_inv
    df.to_csv('../data/stimuli_subject_files/' + subj_file_names[sub_count-1], index=False)
    print(arrayed_current_list_sorted_inv[:1])
    print(subj_file_names[sub_count-1])
    
    print('-----')
```
